```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
df.head()
```

```python
df.columns
```

```python
## Drop day, month & year
df.drop(['day', 'month', 'year'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Encoding
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains("not fire"), 0, 1)
```

```python
df['Classes'].value_counts()
```

```python
# Independent and dependent Features
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
X, y 
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Feature Selection based on correlation
X_train.corr()
```

```python
# Check for Multi-collinearity - when multiple independent features are highly correlated
plt.figure(figsize=(10,6))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
# Returns highly correlated features
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# Threshold - Domain Expert
corr_features = correlation(X_train, 0.85)
```

```python
# Drop highly correlated features
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

### Feature scaling or standardisation

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
X_train_scaled
```

```python
# Box plot to understand the effect of standardScalar
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Model Training

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
y_pred = linreg.predict(X_test_scaled)
```

```python
## Performance Metrices 
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae, r2
```

```python
plt.scatter(y_test, y_pred)
```

### Lasso Regression Model

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, r2_score

lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
```

```python
## Performance Metrices 
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae, r2
```

```python
plt.scatter(y_test, y_pred)
```

### Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
```

```python
## Performance Metrices 
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae, r2
```

```python
plt.scatter(y_test, y_pred)
```

### Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error, r2_score

elastic = ElasticNet()
elastic.fit(X_train_scaled, y_train)
y_pred = elastic.predict(X_test_scaled)
```

```python
## Performance Metrices 
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae, r2
```

```python
plt.scatter(y_test, y_pred)
```

### Cross Validation using LassoCV

```python
from sklearn.linear_model import LassoCV
lascv = LassoCV(cv=5)
lascv.fit(X_train_scaled, y_train)
```

```python
lascv.alpha_
```

```python
y_pred = lascv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
```
